```python
import pandas as pd 
"""
Scheme of work the function ''University_programs()'' return the programs that correspond to each university, and ''top_programs()'' return the top of this program in the country.... 
"""
"""
The next step is store who are the better programs to run table 1...
"""

def University_programs(df:pd.DataFrame, university:str, var_institution:str,var_program_name:str ) -> list :
    df_ = df.copy()
    df_ = df[df[var_institution]==university]
    programs = [x for x in df_[var_program_name].unique() ]
    return programs

def top_programs(df:pd.DataFrame, var_institution: str, var_program_name:str, var_filter: str, field: str,  k:int, university:str) -> pd.DataFrame:
    data = df.copy()
    data = data[data[var_program_name]==var_filter]
    gp = data.groupby(by=[var_institution])
    data[field+' mean']= gp[field].transform('mean')
    data['number of students'] = gp[field].transform('count')
    data.drop_duplicates(subset=var_institution, inplace=True)
    data.sort_values(by=[field+' mean'], ascending=False, inplace=True)
    data =  data[[var_institution,var_program_name,field+' mean']].reset_index(drop=True)
    ranking = data.copy()
    data = data[0:k+1]
    N = ranking.shape[0]
    mean_total = ranking[field+ ' mean'].mean()
    for i in ranking.index:
        if ranking.loc[i,var_institution]==university:
            position = i
    if position > k:
        row = ranking[ranking[var_institution]==university]
        data= data.append(row)
        data = data.append({var_institution:'Total, number and mean',var_program_name:f'{N} programs',  field + ' mean':mean_total} ,ignore_index=True)
        data.rename({data.index[-1]: 'All'}, inplace=True)
        return data
    else:
        data = data.append({var_institution:'Total, number and mean',var_program_name:f'{N} programs',  field + ' mean':mean_total} ,ignore_index=True)
        data.rename({data.index[-1]: 'All'}, inplace=True)
        return data
```

```python
# Analysis of functions
# We need a function that take the top 5 universities in the country and select as reference...
# after take percentiles, and compare the number of individuals in each percentile.
## Information ids, and variables:
"""
INST_COD_INSTITUCION --- código institución
INST_NOMBRE_INSTITUCION --- Nómbre de IES
ESTU_PRGM_ACADEMICO -- Nombre del programa académico
ESTU_SNIES_PRGMACADEMICO -- Código SNIES del programa académico
GRUPOREFERENCIA -- grupo de referencia.
MOD_RAZONA_CUANTITAT_PUNT --- rango de 0 a 300
"""
```

```python
def main():
    ##### Defining globals variables to execution of program.
    var_institution = 'INST_NOMBRE_INSTITUCION'
    var_program_name = 'ESTU_PRGM_ACADEMICO'
    field = 'MOD_RAZONA_CUANTITAT_PUNT'  # test name
    k = 5 # number of program to top
    df = df = pd.read_csv('Genericas2020.txt', sep='¬', encoding='utf8', quotechar="'")
    Universities = ['UNIVERSIDAD CENTRAL-BOGOTÁ D.C.', 'UNIVERSIDAD SURCOLOMBIANA-NEIVA']
    for university in Universities:
        print(university)
        for program in University_programs(df, university, var_institution, var_program_name):
            print(program)
            table = top_programs(df, var_program_name=var_program_name, var_filter=program, field=field, var_institution=var_institution, k=k, university=university)
            display(table)
main()
```
